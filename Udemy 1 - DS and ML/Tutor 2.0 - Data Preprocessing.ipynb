### Data Cleaning - Missing Values

```python
# Missing Values and Data Cleaning
"""
NaN - Not a Number
Null - No Value.
"""
Data_Set6.info()
```

```python
# Import numpy to replace missing value

import numpy as np

Data_Set7 = Data_Set6.replace('!',np.NaN) # to replace missing value with Nan
Data_Set7 = Data_Set7.apply(pd.to_numeric) # To change dataFrame into a numerical data set
Data_Set7.info()
```

```python
# to spot a null value from data set where need to be clean
Data_Set7.isnull() 
```

```python
# To drop all missing values. 1st Option
Data_Set7.dropna(axis=0,inplace=True)
Data_Set7
```

```python
# To replace missing values. 2nd Option
Data_Set8 = Data_Set7.fillna(method='ffill') # using before data to replace a nan value
Data_Set8
```

```python
# To replace missing values. 2nd Option
from sklearn.impute import SimpleImputer 
M_var = SimpleImputer(missing_values=np.nan, strategy='mean') #To replace missing value with mean value
M_var.fit(Data_Set7)
Data_Set9 = M_var.transform(Data_Set7)
Data_Set9 = pd.DataFrame(Data_Set9, columns = ['Time','E-Plug','E-Heat','Price','Temp'])
Data_Set9 = Data_Set9.round()
Data_Set9
```

### Concatenation

```python
# Concatenation
New_Col = pd.read_csv('Data_New.csv')
Data_Set10 = pd.concat([Data_Set8,New_Col],axis=1) # to combine to data coloums
Data_Set10
```

### Dummy Coding

```python
# Changing Categorical variable into Numeriacal

Data_Set11 = pd.get_dummies(Data_Set10) # Easy Way
Data_Set11.rename(columns={ 'P/OffP _OffPeak': 'OffPeak','P/OffP _Peak':'Peak' }, inplace =True)
Data_Set11
```

### Normalization

```python
# Importing minmax normalize in sklearn
from sklearn.preprocessing import minmax_scale, normalize
```

```python
# First Method: Min Max Scale
Data_Set12 = minmax_scale(Data_Set11, feature_range=(0,1))
print(Data_Set12)
```

```python
# axis = 0 for normalizing features / axis = 1 is for normalizing each sample.
Data_Set13 = normalize(Data_Set11, norm= 'l2',axis=0)
print(Data_Set13.round(2))
```

```python
Data_Set13 = pd.DataFrame(Data_Set13,columns=['Time','E-Plug','E-Heat','Price','Temp','OffPeak','Peak'])
Data_Set13.round(3) #Last Step on Data preprocessing Done.
```

## Clustering Techniques

```python
# importing modules
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating Data Set
iris = load_iris()
```

```python
Data_iris = iris.data
```

### K-Mean Clustring

```python
# Importing K-mean from sklearn cluster
from sklearn.cluster import KMeans
KMNS = KMeans(n_clusters=3)
```

```python
KMNS.fit(Data_iris)
```

```python
# Predicting model
labels = KMNS.predict(Data_iris)
```

```python
Ctn = KMNS.cluster_centers_
```

```python
# Plotting Model
plt.scatter(Data_iris[:,2],Data_iris[:,3],c = labels)
plt.scatter(Ctn[:,2],Ctn[:,3], marker = 'o',color='red',s = 120)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

```

```python
# Inertia for the model
KMNS.inertia_
```

```python
# Creating Indertia empty list
K_inertia = []
```

```python
for i in range(1,10):
    KMNS = KMeans(n_clusters=i, random_state=44)
    KMNS.fit(Data_iris)
    K_inertia.append(KMNS.inertia_)
```

```python
#plotting all the KMNS inertia
plt.plot(range(1,10),K_inertia,color = 'green',marker = 'v')
plt.xlabel('Number of K')
plt.ylabel('Inertia')
plt.show()
```

Based on the rule of elbow method the best number of k is 3

### DBSCAN

```python
# importing DBSCAN from sklearn cluster
from sklearn.cluster import DBSCAN
DBS = DBSCAN(eps = 1,min_samples= 4)
```

```python
DBS.fit(Data_iris)
Labels = DBS.labels_
Labels
```

```python
plt.scatter(Data_iris[:,2],Data_iris[:,3],c = Labels)
plt.show()
```

### Hierarchical Clustring

```python
# Importing Hierarchical cluster from sklearn
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
HR = linkage(Data_iris, method = 'complete')
```

```python
Dnd = dendrogram(HR)
```

```python
Labels = fcluster(HR,4, criterion= 'distance')
```

```python
plt.scatter(Data_iris[:,2],Data_iris[:,3] , c=Labels)
plt.show()
```

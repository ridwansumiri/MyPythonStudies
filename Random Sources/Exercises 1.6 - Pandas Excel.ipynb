```python
"""
23. Write a Pandas program to import excel data (coalpublic2013.xlsx ) 
into a dataframe and draw a bar plot where each bar will represent
one of the top 10 production
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel('E:\Python Data\Data\coalpublic2013.xlsx')
sorted_by_production = df.sort_values(['Production'], ascending=False).head(10)
sorted_by_production['Production'].head(10).plot(kind="barh")
plt.show()
```
